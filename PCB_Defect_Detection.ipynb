```python
import cv2
import matplotlib.pyplot as plt
import glob
import os

predicted_images_path = "/content/runs/detect/predict"
image_paths = glob.glob(os.path.join(predicted_images_path, "*.jpg"))

print(f"Found {len(image_paths)} images.")

if len(image_paths) == 0:
    print("No images found in the prediction folder.")
else:
    num_images_to_show = min(5, len(image_paths))
    selected_images = image_paths[:num_images_to_show]

    fig, axes = plt.subplots(1, num_images_to_show, figsize=(15, 5))

    if num_images_to_show == 1:
        axes = [axes]

    for i, img_path in enumerate(selected_images):
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        axes[i].imshow(img)
        axes[i].axis("off")

    plt.show()
```
